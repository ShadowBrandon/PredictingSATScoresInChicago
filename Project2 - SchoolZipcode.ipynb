**GOAL:** From a data set about ALL schools from Chicago, I want a new data frame with school, type, zip, lat, long and EXPORT THAT DATA AS A SPREADSHEET!


```python
#Goal: Go through the type and find High School Charter to High School

dfs = pd.read_excel("SchoolDataMerged_August2017.xlsx")
dfs['Type'] = dfs['Type'].replace(['High School Charter', 'High School'], 'High School')
dfs['School'] = dfs['School'].str.replace(' HS', '')

dfs['Type'].head(3)
dfs['School'].head(40)

#df['range'].replace(',','-',inplace=True)
```

```python
dfs.info()
```

```python
#Goal: make a new data frame with school, type, zip, lat, long

new_dfs = dfs.loc[dfs['Type'] == "High School"]
#new_dfs.head(20)
new_dfs = new_dfs[["School","Type","Zip","Lat","Long"]].reset_index(drop=True)
#new_dfs.head(20)
```

```python
new_dfs.to_csv("SimplifiedSchoolDataMerged_August2017.csv")
```
